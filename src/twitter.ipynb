```
from twitterscraper import query_tweets
```

```
from datetime import datetime
#file = open("output.txt","w")
twitter=[]
query_res = query_tweets("EmeryOut",  50000, begindate=datetime.strptime('2019-11-11', '%Y-%m-%d').date())[:50000]
for tweet in query_res:
    #print(tweet.text)
    twitter.append(tuple(("EmeryOut",str(tweet.text),tweet.replies,tweet.retweets,tweet.likes,    tweet.timestamp)))
    #print(tweet.text)
print(len(twitter))
```

```
from datetime import datetime
query_res1 = query_tweets("PochettinoOut",  50000, begindate=datetime.strptime('2019-8-10', '%Y-%m-%d').date())[:50000]
query_res2 = query_tweets("PochOut",  50000, begindate=datetime.strptime('2019-8-10', '%Y-%m-%d').date())[:50000]
```

```
twitter_pochettino = []
twitter_pochout = []
for tweet in query_res1:
  if tweet not in twitter_pochettino:
    twitter_pochettino.append(tuple(("PochettinoOut",str(tweet.text),tweet.replies,tweet.retweets,tweet.likes,    tweet.timestamp)))
for tweet in query_res2:
  if tweet not in twitter_pochout:
    twitter_pochout.append(tuple(("PochOut",str(tweet.text),tweet.replies,tweet.retweets,tweet.likes,    tweet.timestamp)))
print(len(twitter_pochout))
print(len(twitter_pochettino))
```

```
pochettinooutdf = pd.DataFrame(list(twitter_pochettino), columns=["hashtag","tweet text", "replies","retweets", "likes", "date"])
pochettinooutdf = pochettinooutdf.drop_duplicates()
```

```
pochoutdf = pd.DataFrame(list(twitter_pochout), columns=["hashtag","tweet text", "replies","retweets", "likes", "date"])
pochoutdf = pochoutdf.drop_duplicates()
```

```
pochettinooutdf.to_csv(path_or_buf="./pochettino.tsv", sep="\t", index=False)
```

```
pochoutdf.to_csv(path_or_buf="./pochout.tsv", sep="\t", index=False)
```
